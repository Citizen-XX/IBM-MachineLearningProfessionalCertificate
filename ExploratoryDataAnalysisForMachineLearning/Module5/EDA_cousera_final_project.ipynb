# Outliers

Create a function to find outliers across numerical columns, with keen interest on popularity and identify the size and impact into the dateset.

1. Define a Function to Find Outliers
2. Calculates the first quartile (Q1), third quartile (Q3), and Interquartile Range (IQR) for each numerical column using df[col].quantile().
3. Determines the lower and upper bounds for outliers using Q1 - 1.5 * IQR and Q3 + 1.5 * IQR.
4. Identifies and stores rows that fall outside these bounds in a dictionary.

```python
# List of numerical columns to check for outliers
numerical_columns = ['popularity']
#  , 'acousticness', 'danceability', 'duration_ms', 'energy','instrumentalness', 'liveness', 'loudness', 'speechiness', 'tempo', 'valence']

# Function to find outliers using IQR
def find_outliers_iqr(df, columns):
    outliers = {}
    for col in columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        outliers[col] = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    return outliers

# Find outliers
outliers_iqr = find_outliers_iqr(data, numerical_columns)


# Display outliers for each column
for col, outliers in outliers_iqr.items():
    print(f"Outliers for {col}:\n{outliers}\n")
```

# Feature engineering

Create encode categorical variables e.g genre, key, mode to numerical varibales and new features.

## Create encode categorical variables

```python
# Create a copy of the cleaned data
data_clean = df.copy()
```

```python
# Encoding categorical variables

def encode_categorical_columns(df, columns):
    """
    Encodes categorical columns in a DataFrame using category codes.

    Parameters:
    df (pd.DataFrame): The DataFrame containing the columns to encode.
    columns (list of str): List of column names to encode.

    Returns:
    pd.DataFrame: The DataFrame with encoded columns. """
    for column in columns:
        df[column] = df[column].astype('category').cat.codes
    return df

# List of columns to encode
columns_to_encode = ['key', 'mode', 'time_signature']

# Encode the categorical variables in the DataFrame

data_clean = encode_categorical_columns(data_clean, columns_to_encode)
```

## Create a new feature:

After encoding the categorical variables, create new features on the dataset 
1. Convert the duration_ms column to minutes 
2. Create a binary column for popularity such that if the track scores below 50% it is not popular.

```python
def create_new_feature_duration(df):
    """
    Creates new features in the DataFrame.

    Parameters:
    df (pd.DataFrame): The DataFrame to modify, such that convert to minutes

    Returns:
    pd.DataFrame: The DataFrame with new features added.
    """

    df['duration_minutes'] = df['duration_ms'] / 60000

    return df

def create_new_feature_popularity(df):
    """
    Creates new features in the DataFrame.

    Parameters:
    df (pd.DataFrame): The DataFrame to modify, such that a new feature: 'is_popular' based on the 'popularity' score

    Returns:
    pd.DataFrame: The DataFrame with new features added.

    """

    # Create a new feature: 'is_popular' based on the 'popularity' score
    df['is_popular'] = df['popularity'].apply(lambda x: 1 if x > 50 else 0)

    return df
```

```python
data_clean = create_new_feature_duration(data_clean)
data_clean = create_new_feature_popularity(data_clean)
```

```python
# Calculate genre popularity percentage
genre_popularity_percentage = (data_clean['genre'].value_counts() / len(data_clean)) * 100
top_10_genres = genre_popularity_percentage.head(10)

# Calculate percentage for 'Others' category
other_genres_percentage = genre_popularity_percentage[10:].sum()

# Create a list of genres for the pie chart
genres_list = list(top_10_genres.index)
genres_list.append('Others')

# Create a corresponding list of percentages
percentages_list = list(top_10_genres.values)
percentages_list.append(other_genres_percentage)

# Plotting the pie chart
plt.figure(figsize=(10, 8))
plt.pie(percentages_list, labels=genres_list, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('viridis', len(genres_list)))
plt.title('Top 10 Most Popular Genres on Spotify by %', fontsize=15)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
